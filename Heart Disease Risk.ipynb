```python
#Model building
Ind_features = list(data.columns)
Ind_features.remove('target')

x = data[Ind_features] #features
y = data['target'] #target variable
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.30, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
#####Model Fitting/Training###
lr.fit(x_train, y_train)
```

```python
test_pred = lr.predict(x_test)
```

```python
test_pred
```

```python
pred = pd.DataFrame() 
pred["y_test"] = y_test #actual answer key
pred["y_pred"] = test_pred #answers given by model
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,test_pred)
```

```python
# Therefore, the accuracy of the model is 78.3018
```
